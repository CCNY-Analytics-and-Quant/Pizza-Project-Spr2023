```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Resultsd.csv')
```

```python
df['revenue'] = df['price']
```

```python
grouped_df = df.groupby('pizza_id').agg({'revenue': 'sum'})
```

```python
top_pizzas = grouped_df.sort_values('revenue', ascending=False).head(10)
```

```python
sns.barplot(x=top_pizzas.index, y='revenue', data=top_pizzas)
plt.title('Revenue by Pizza Type')
plt.xlabel('Pizza Type')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=90) 
plt.savefig('CHART5.png', dpi=300, bbox_inches='tight')
```
